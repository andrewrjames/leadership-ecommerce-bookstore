### Book Genres Bought Together

```python
columns = [col for col in df_new.columns if col.startswith('frequency') or col.startswith('id')]
df_v2= df_new.copy()
df_v2
for col in columns[1:]:
  df_v2[col] = np.where(df_v2[col] >= 1, 1, 0)

df_final = df_v2[columns]
df_final = df_final.melt(value_vars = columns[1:], var_name='book_genre', id_vars='id', value_name='frequency')
df_final['book_genre'] = df_final['book_genre'].astype(str)
df_final['book_genre'] = df_final['book_geånre'].str.extract(r'frequency\_(.*)')
df_final.head()
# df_final.head(5)
# df_new.melt(value_name = columns, id_vars='id', var_name='book_genre', '')
```

```python
final = df_final.pivot(index='id', columns='book_genre', values='frequency')
co_occurrence_matrix = final.T.dot(final)
# np.fill_diagonal(co_occurrence_matrix.values, 0) # art vs art will be set to zero
```

```python
co_occurrence_matrix_sqrt = np.sqrt(co_occurrence_matrix)
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(co_occurrence_matrix_sqrt, annot=False, cmap="Blues", fmt="d")
```

```python
pd.reset_option('all')
association_matrix = df_v2[columns]
association_matrix.drop('id', axis=1, inplace=True)
association_matrix
# association_matrix.set_index('id', inplace=True)
# association_matrix = association_matrix.reset_index()
```

```python
# !pip install --upgrade mlxtend
from mlxtend.frequent_patterns import apriori, association_rules

# Generate frequent itemsets
frequent_itemsets = apriori(association_matrix, min_support=0.1, use_colnames=True)

num_itemsets = len(association_matrix)
# Generate association rules
rules = association_rules(frequent_itemsets, num_itemsets=num_itemsets, metric="lift", min_threshold=1.0)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
rules.sort_values('confidence', ascending=False)
```

### Rule-based segmentation using RFM

```python
df_new['r_quartile'] = pd.qcut(df_new['r'], 4, labels=['1','2','3','4'])
df_new['f_quartile'] = pd.qcut(df_new['f'], 4, labels=['4','3','2','1'])
df_new['m_quartile'] = pd.qcut(df_new['m'], 4, labels=['4','3','2','1'])
```

```python
df_new.groupby('r_quartile').agg({'id':'count'}).reset_index().rename(columns={'id':'num_customers'})
```

```python
df_new.groupby('f_quartile').agg({'id':'count'}).reset_index().rename(columns={'id':'num_customers'})
```

```python
df_new.groupby('m_quartile').agg({'id':'count'}).reset_index().rename(columns={'id':'num_customers'})
```

```python
df_new['fm_score'] = df_new['f_quartile'].astype(int) + df_new['m_quartile'].astype(int)
```

```python
df_new[['r_quartile','fm_score']].value_counts().reset_index().sort_values('fm_score', ascending=True)
```

```python
df_new['r_quartile'] = df_new['r_quartile'].astype(int)

conditions = [(df_new['fm_score'].isin([2,3])) &  (df_new['r_quartile'].isin([1,2])), # champion
              (df_new['fm_score'].isin([2,3])) &  (df_new['r_quartile'].isin([3])), # potential loyalist
              (df_new['fm_score'].isin([2,3])) &  (df_new['r_quartile'].isin([4])), # hibernating champion
              (df_new['fm_score'].isin([4,5,6])) &  (df_new['r_quartile'].isin([1,2])), # recent customers
              (df_new['fm_score'].isin([4,5,6])) &  (df_new['r_quartile'].isin([3,4])), # long lost customers
              (df_new['fm_score'].isin([7,8])) &  (df_new['r_quartile'].isin([1,2])), # recent customers
              (df_new['fm_score'].isin([7,8])) &  (df_new['r_quartile'].isin([3,4]))] # long lost customers

values = ['01. champion',
          '02. potential loyalist',
          '04. hibernating champion',
          '03. recent customers',
          '05. long lost customers',
          '03. recent customers',
          '05. long lost customers']

df_new['rfm_segment'] = np.select(conditions, values, default= 'Unknown')
```

```python
square = df_new['rfm_segment'].value_counts().reset_index()
square
```

```python
summary = df_new.groupby('rfm_segment').describe().T
summary = summary.loc[['r', 'f', 'm']]
summary.T
```
